# Diccionario de Terminologías Médicas

| **Término**                                | **Definición**                                                                                                                                                                                                                                       |
| ------------------------------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Angina**                                 | Dolor o molestia en el pecho causada por una disminución del flujo sanguíneo al corazón. Puede ser *típica* (con patrón predecible) o *atípica* (síntomas menos claros).                                                                             |
| **Angina inducida por ejercicio (Exang)**  | Episodio de dolor torácico que aparece al realizar esfuerzo físico, debido a una insuficiencia en el flujo sanguíneo cardíaco durante la actividad.                                                                                                  |
| **Colesterol (Chol)**                      | Sustancia grasa que circula por la sangre. Niveles altos pueden provocar acumulación de placas en las arterias y aumentar el riesgo de enfermedad cardíaca.                                                                                          |
| **Depresión del segmento ST (Oldpeak)**    | Alteración en el electrocardiograma que indica posible isquemia cardíaca (falta de oxígeno en el corazón).                                                                                                                                           |
| **Electrocardiograma en reposo (RestECG)** | Prueba que registra la actividad eléctrica del corazón mientras el paciente está en reposo. Permite detectar arritmias o lesiones cardíacas.                                                                                                         |
| **Fluoroscopía (CA)**                      | Técnica de rayos X en tiempo real que permite observar el funcionamiento del corazón y vasos sanguíneos; en este caso, mide cuántos vasos principales presentan flujo anormal.                                                                       |
| **Glucemia en ayunas (FBS)**               | Nivel de azúcar en la sangre tras un periodo sin ingerir alimentos. Valores elevados pueden indicar riesgo de diabetes o síndrome metabólico.                                                                                                        |
| **Índice de Masa Corporal (IMC)**          | Medida que relaciona peso y altura para evaluar si una persona tiene un peso saludable. Aunque no está en la tabla original, suele incluirse como variable complementaria.                                                                           |
| **Presión arterial en reposo (Trestbps)**  | Fuerza que ejerce la sangre contra las paredes de las arterias cuando el corazón está en reposo entre latidos.                                                                                                                                       |
| **Segmento ST**                            | Parte del electrocardiograma que representa el periodo entre la despolarización y repolarización del ventrículo; cambios en su forma indican daño o estrés cardíaco.                                                                                 |
| **Talio / Thalassemia (Thal)**             | En este contexto se refiere a un *test de talio*, una prueba de esfuerzo con un isótopo radiactivo que evalúa el flujo sanguíneo al corazón. Los resultados pueden indicar defectos fijos (daño permanente) o reversibles (falta temporal de flujo). |
| **Pendiente del ST (Slope)**               | Describe la inclinación del segmento ST durante el esfuerzo. Su forma puede indicar si el corazón recibe suficiente oxígeno durante el ejercicio.                                                                                                    |
| **Frecuencia cardíaca máxima (Thalach)**   | El número más alto de latidos por minuto que una persona alcanza durante una prueba de esfuerzo. Es un indicador de la capacidad cardiovascular.                                                                                                     |
| **Vasos principales (CA)**                 | Arterias coronarias más grandes que suministran sangre al corazón; su bloqueo o estrechamiento puede causar angina o infarto.                                                                                                                        |

## ¿La hipertensión se asocia con la presencia de enfermedad cardíaca?



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import chi2_contingency

# 1Crear variable categórica 'hipertenso'
df['hipertenso'] = (df['trestbps'] > 130).astype(int)  # 1 = hipertenso, 0 = no hipertenso

# Ver distribución general
print(df['hipertenso'].value_counts())

# Tabla de contingencia
tabla = pd.crosstab(df['hipertenso'], df['target'])
print("\nTabla de contingencia:")
print(tabla)

# Prueba Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla)
print(f"\nChi-cuadrado = {chi2:.2f}")
print(f"p-valor = {p:.4f}")

if p < 0.05:
    print(" Existe una asociación significativa entre hipertensión y enfermedad cardíaca.")
else:
    print(" No se encontró una asociación estadísticamente significativa entre hipertensión y enfermedad cardíaca.")

# Gráfico comparativo
sns.countplot(data=df, x='hipertenso', hue='target', palette=['#8FD694', '#E57373'])
plt.title('Relación entre hipertensión y enfermedad cardíaca')
plt.xlabel('Hipertenso (1 = Sí, 0 = No)')
plt.ylabel('Cantidad de pacientes')
plt.legend(title='Enfermedad', labels=['No','Sí'])
plt.show()

```

Se encontró una asociación significativa (p = 0.0058) entre hipertensión y enfermedad cardíaca, lo que sugiere que la presión arterial elevada podría influir en el riesgo cardiovascular dentro de esta muestra.

# Correlación entre variables.



### **Análisis de Correlación entre Variables**

Para obtener una visión general de las relaciones lineales entre las distintas variables numéricas del conjunto de datos, generamos una **matriz de correlación**. La matriz se visualiza con un  **mapa de calor (heatmap)**, una herramienta gráfica que utiliza una escala de colores para representar la fuerza y la dirección de las correlaciones.

En este mapa de calor:
*   Los colores más **claros o cálidos** indican una **correlación positiva fuerte** (cuando una variable aumenta la otra tambien).
*   Los colores más **oscuros o fríos** indican una **correlación negativa fuerte** (cuando el valor una variable aumenta la otra debe disminuir).
*   Los colores **intermedios** sugieren una **correlación débil o nula**.

Nuestro principal interés se centra en la última fila/columna, que muestra la correlación de cada variable con nuestra variable objetivo (`target`), dándonos pistas iniciales sobre los factores más predictivos. Adicionalmente, este análisis nos permite detectar posible **multicolinealidad** (alta correlación entre variables predictoras).


```python

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr())
plt.title('Mapa de Calor de Correlación entre Variables', fontsize=16)
plt.show()
```


### **Análisis del Gráfico de Correlación**

El mapa de calor nos permitio identificar:

**1. Correlaciones con la Variable Objetivo:**

*   **Correlaciones Positivas más Fuertes:**
    *   **Tipo de dolor de pecho(cp):** Con una correlación de **0.43**, es uno de los predictores positivos más fuertes. Esto indica que ciertos tipos de dolor de pecho (específicamente los valores 1, 2 y 3) están asociados con una mayor probabilidad de tener una enfermedad cardíaca.

    *   **Pendiente del segmento ST(slope):** Con una correlación de **0.35**, sugiere que una pendiente ascendente (valor 2) está ligada a un diagnóstico positivo.

*   **Correlaciones Negativas más Fuertes:**
    *   **Angina inducida por ejercicio(exang):** Tiene una correlación de **-0.44**. Esto significa que la presencia de angina durante el ejercicio (`exang=1`) está fuertemente asociada con la **ausencia** de enfermedad cardíaca.
    *   **Depresión del ST(oldpeak):** Con una correlación de **-0.43**, valores más altos de depresión del ST se asocian con una menor probabilidad de enfermedad.
    *   **Número de vasos coloreados(ca):** Muestra una correlación de **-0.39**. A medida que aumenta el número de vasos con obstrucción, la probabilidad de tener la enfermedad disminuye, confirmando la relación inversa que vimos en el análisis bivariado.
    *   **Sexo(sex):** Con **-0.28**, indica que ser hombre (`sex=1`) está asociado a una menor probabilidad de enfermedad en este dataset.

*   **Correlaciones Débiles:**

**2. Correlación entre Variables Predictoras (Multicolinealidad):**

El mapa de calor confirma de manera cuantitativa los hallazgos del análisis bivariado. Las variables más prometedoras para predecir la enfermedad cardíaca son `cp`, `thalach`, `slope`, `exang`, `oldpeak` y `ca`. Por otro lado, factores de riesgo clásicos como el colesterol y la presión arterial en reposo no muestran una fuerte relación lineal con el diagnóstico en esta muestra.

# Modulo de Recomendaciones

## 1. Lógica del Sistema de Recomendaciones

```python
import gradio as gr
import pandas as pd

# Mapeo de etiquetas legibles a los valores numéricos del dataset
SEX_MAP = {"Mujer (0)": 0, "Hombre (1)": 1}
CP_MAP = {"0 - Angina típica": 0, "1 - Angina atípica": 1, "2 - Dolor no anginoso": 2, "3 - Asintomático": 3}
FBS_MAP = {"0 - Normal (≤120 mg/dl)": 0, "1 - Alta (>120 mg/dl)": 1}
RESTECG_MAP = {"0 - Normal": 0, "1 - Anomalía ST-T": 1, "2 - HVI probable": 2}
EXANG_MAP = {"0 - No": 0, "1 - Sí": 1}
SLOPE_MAP = {"0 - Descendente": 0, "1 - Plano": 1, "2 - Ascendente": 2}
THAL_MAP = {"0 - No aplica": 0, "1 - Defecto fijo": 1, "2 - Defecto reversible": 2, "3 - Normal": 3}

def recomendar_salud_cardiaca(age, sex_lbl, cp_lbl, trestbps, chol, fbs_lbl, thalach, exang_lbl, oldpeak, slope_lbl, thal_lbl):
    """
    Calcula un puntaje de riesgo y genera recomendaciones basadas en guías médicas generales
    y los hallazgos del análisis del dataset.
    """
    risk_score = 0
    recommendations = []

    # Mapear entradas de texto a valores numéricos
    sex = SEX_MAP[sex_lbl]
    cp = CP_MAP[cp_lbl]
    fbs = FBS_MAP[fbs_lbl]
    exang = EXANG_MAP[exang_lbl]
    slope = SLOPE_MAP[slope_lbl]
    thal = THAL_MAP[thal_lbl]

    # --- Lógica de Puntuación de Riesgo (1 punto por cada factor de riesgo) ---

    # 1. Edad y Sexo
    if (sex == 1 and age > 45) or (sex == 0 and age > 55):
        risk_score += 1
        recommendations.append("• **Edad/Sexo:** La edad es un factor de riesgo natural. Es importante mantener controles médicos regulares, especialmente a partir de los 45 años para hombres y 55 para mujeres.")

    # 2. Presión Arterial (trestbps)
    if trestbps > 130:
        risk_score += 1
        recommendations.append(f"• **Presión Arterial ({trestbps} mm Hg):** Tu presión arterial en reposo está por encima del nivel óptimo (130 mm Hg). Considera reducir el consumo de sal, hacer ejercicio regularmente y consultar a un médico.")

    # 3. Colesterol (chol)
    if chol > 240:
        risk_score += 1
        recommendations.append(f"• **Colesterol ({chol} mg/dl):** Tu nivel de colesterol es alto (superior a 240 mg/dl). Una dieta baja in grasas saturadas y rica en fibra puede ayudar a controlarlo.")

    # 4. Angina Inducida por Ejercicio (exang)
    if exang == 1:
        risk_score += 1
        recommendations.append("• **Angina por Ejercicio:** Experimentar dolor de pecho durante el ejercicio es una señal de alerta importante. Es fundamental que consultes a un cardiólogo.")

    # 5. Depresión del ST (oldpeak)
    if oldpeak > 1.0:
        risk_score += 1
        recommendations.append(f"• **Depresión del ST ({oldpeak}):** Un valor de 'oldpeak' elevado sugiere una respuesta anormal del corazón al ejercicio. Este es un indicador que tu médico debe evaluar en detalle.")

    # 6. Frecuencia Cardíaca Máxima (thalach)
    predicted_max_hr = 220 - age
    if thalach < (predicted_max_hr * 0.85): # Si es menor al 85% de la esperada
        risk_score += 1
        recommendations.append(f"• **Frecuencia Cardíaca Máxima ({thalach} lpm):** Tu frecuencia cardíaca máxima es más baja de lo esperado para tu edad. Esto puede indicar una menor capacidad cardiovascular. El ejercicio aeróbico regular puede ayudar a mejorarla.")

    # 7. Test de Talio (thal)
    if thal == 2: # Defecto reversible
        risk_score += 2 # Se le da más peso por ser un predictor fuerte
        recommendations.append("• **Test de Talio (Defecto Reversible):** Este resultado es un indicador muy fuerte de problemas de flujo sanguíneo al corazón. Es crucial que un especialista evalúe este hallazgo.")
    elif thal == 1: # Defecto fijo
        risk_score += 1
        recommendations.append("• **Test de Talio (Defecto Fijo):** Este resultado puede indicar una cicatriz de un evento cardíaco previo. Es importante un seguimiento médico continuo.")


    # --- Determinación del Nivel de Riesgo ---
    if risk_score <= 2:
        risk_level = "Riesgo Bajo"
        if not recommendations:
            recommendations.append("¡Felicidades! Tus indicadores se encuentran dentro de rangos saludables. Continúa con un estilo de vida activo y una dieta balanceada.")
    elif risk_score <= 4:
        risk_level = "Riesgo Moderado"
        recommendations.insert(0, "Se han identificado algunos factores de riesgo. Es un buen momento para adoptar hábitos más saludables y considerar una visita preventiva al médico.")
    else:
        risk_level = "Riesgo Alto"
        recommendations.insert(0, "Se han identificado múltiples factores de riesgo importantes. Te recomendamos encarecidamente que consultes a un médico o cardiólogo para una evaluación completa.")

    return risk_level, "\n\n".join(recommendations)
```

##2. Interfaz Gráfica con Gradio

```python
# Creación de la interfaz de Gradio
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# Predictor de Riesgo Cardíaco y Recomendaciones")
    gr.Markdown(
        "Introduce tus datos para obtener una estimación de tu riesgo cardiovascular y recibir recomendaciones personalizadas. "
        "**Aviso:** Esta es una herramienta educativa y no reemplaza el diagnóstico médico profesional."
    )

    with gr.Row():
        with gr.Column():
            gr.Markdown("### Datos Fisiológicos")
            age_input = gr.Slider(29, 80, value=54, label="Edad (años)")
            trestbps_input = gr.Slider(90, 200, value=130, label="Presión Arterial en Reposo (mm Hg)")
            chol_input = gr.Slider(120, 600, value=240, label="Colesterol (mg/dl)")
            thalach_input = gr.Slider(70, 210, value=150, label="Frecuencia Cardíaca Máxima")
            oldpeak_input = gr.Slider(0.0, 6.5, value=1.0, step=0.1, label="Depresión del ST (oldpeak)")

        with gr.Column():
            gr.Markdown("### Historial y Pruebas")
            sex_input = gr.Dropdown(list(SEX_MAP.keys()), value="Hombre (1)", label="Sexo")
            cp_input = gr.Dropdown(list(CP_MAP.keys()), value="2 - Dolor no anginoso", label="Tipo de Dolor de Pecho")
            fbs_input = gr.Dropdown(list(FBS_MAP.keys()), value="0 - Normal (≤120 mg/dl)", label="Glucemia en Ayunas")
            exang_input = gr.Dropdown(list(EXANG_MAP.keys()), value="0 - No", label="Angina Inducida por Ejercicio")
            slope_input = gr.Dropdown(list(SLOPE_MAP.keys()), value="1 - Plano", label="Pendiente del Segmento ST")
            thal_input = gr.Dropdown(list(THAL_MAP.keys()), value="2 - Defecto reversible", label="Resultado del Test de Talio")
            restecg_input = gr.Dropdown(list(RESTECG_MAP.keys()), visible=False) # Variable no usada en la lógica, pero necesaria para la función

    submit_btn = gr.Button("Analizar Riesgo", variant="primary")

    with gr.Row():
        risk_output = gr.Textbox(label="Nivel de Riesgo Estimado")
        recommendations_output = gr.Textbox(label="Recomendaciones Personalizadas", lines=10)

    submit_btn.click(
        fn=recomendar_salud_cardiaca,
        inputs=[
            age_input, sex_input, cp_input, trestbps_input, chol_input, fbs_input,
            thalach_input, exang_input, oldpeak_input, slope_input, thal_input
        ],
        outputs=[risk_output, recommendations_output]
    )

demo.launch(debug=True)
```
