# PILA

La analog√≠a m√°s com√∫n y f√°cil de entender para una pila (stack) es una:
## Pila de platos. üçΩÔ∏è

Imagina que est√°s en una cocina y apilas los platos limpios uno encima del otro.

Cuando pones un plato nuevo, lo colocas en la parte superior de la pila.

Cuando necesitas un plato, lo tomas del tope de la pila.

Esto ilustra perfectamente el principio LIFO (Last-In, First-Out): el √∫ltimo plato que se puso es el primero que se saca. No puedes acceder f√°cilmente a los platos de abajo sin quitar primero los que est√°n encima.

````python
stack = []

stack.append(1)
stack.append(2)
stack.append(3)

print(stack)
````

## M√©todos Principales:
1. push()
Este m√©todo se usa para agregar un nuevo elemento a la parte superior de la pila. Es la operaci√≥n de "apilar" un nuevo √≠tem.

Analog√≠a: Poner un nuevo libro encima de una pila de libros.

2. pop()
Este m√©todo elimina y devuelve el elemento que est√° en la cima de la pila. Es el m√©todo que respeta la regla LIFO.

Analog√≠a: Tomar el libro que est√° hasta arriba de la pila para leerlo.

3. peek() (o top())
Este m√©todo devuelve el elemento en la cima de la pila sin eliminarlo. Es √∫til para "echar un vistazo" al siguiente elemento sin modificar la pila.

Analog√≠a: Mirar la portada del libro que est√° en la parte superior sin levantarlo de la pila.

## COLA

Una cola funciona de manera opuesta a una pila. Mientras que en una pila solo se puede acceder al √∫ltimo elemento, en una cola el acceso se realiza por ambos extremos: uno para agregar elementos y otro para eliminarlos.

````python
# Creamos una cola vac√≠a usando una lista
cola_sencilla = []

# Agregamos elementos a la cola (operaci√≥n 'enqueue')
cola_sencilla.append("elemento_1") #1ero
cola_sencilla.append("elemento_2") #2do
cola_sencilla.append("elemento_3") #3ro

print("Cola despu√©s de agregar elementos:", cola_sencilla)

# Eliminamos el primer elemento de la cola (operaci√≥n 'dequeue')
primer_elemento = cola_sencilla.pop()

print("\nElemento eliminado:", primer_elemento)
print("Cola despu√©s de eliminar:", cola_sencilla)
````

## Operaciones Principales

#### Encolar (Enqueue):
La operaci√≥n de agregar un elemento al final de la cola.


#### Desencolar (Dequeue):
La operaci√≥n de eliminar el elemento que est√° al principio de la cola.

````python
from collections import deque

# 1. Crear una cola
cola = deque()

print("La cola est√° vac√≠a:", len(cola) == 0)

# 2. Encolar (enqueue): agregar elementos al final
print("\n--- Agregando elementos a la cola (enqueue) ---")
cola.append("Tarea A")
cola.append("Tarea B")
cola.append("Tarea C")
print("Cola actual:", cola)

# 3. Desencolar (dequeue): eliminar el elemento del frente
print("\n--- Eliminando elementos de la cola (dequeue) ---")
tarea_atendida = cola.popleft()  # popleft() elimina el primer elemento
print("Elemento atendido:", tarea_atendida)
print("Cola despu√©s del primer dequeue:", cola)

tarea_atendida = cola.popleft()
print("Elemento atendido:", tarea_atendida)
print("Cola despu√©s del segundo dequeue:", cola)

# 4. Estado de la cola
print("\n--- Estado final ---")
print("La cola est√° vac√≠a:", len(cola) == 0)
````

````python
from collections import deque

# Creamos una instancia de deque
cola_profesional = deque()

# Encolar: agregamos elementos a la derecha
cola_profesional.append("cliente_A")
cola_profesional.append("cliente_B")
cola_profesional.append("cliente_C")
print(f"Cola de clientes: {cola_profesional}")

# Desencolar: quitamos el elemento de la izquierda
cliente_atendido = cola_profesional.popleft()
print(f"Cliente atendido: {cliente_atendido}")
print(f"Cola actual: {cola_profesional}")
````

## ARBOLES BINARIOS

Un √°rbol binario es una estructura de datos jer√°rquica que se asemeja a un √°rbol geneal√≥gico. Cada nudo en el √°rbol tiene un valor y puede tener, como m√°ximo, dos "hijos": un hijo izquierdo y un hijo derecho.

image.png

A diferencia de las estructuras de datos lineales como las listas,las pilas y las colas,los √°rboles organizan los datos de forma jer√°rquica.Esto los hace muy eficientes para buscar,Insertar y eliminar elementos.La base de un √°rbol es sura√≠z,el primer nodo del cual se desprenden todos los dem√°s.Los nudos que no tienen hijos se llamanhojas.



```
# En Python,no existe un tipo de dato "√°rbol binario" incorporado. Por lo tanto, se suele definir con una clase:

class Nodo

que contiene tres atributos:

valor:El dato que almacena el nodo.

izquierda:Una referencia al nodo hijo izquierdo.

derecha:Una referencia al nodo hijo derecho.
```



````python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None  # Hijo izquierdo
        self.derecha = None    # Hijo derecho
````

### **Recorridos**
Estos son los m√©todos m√°s importantes y se basan en un enfoque de recursi√≥n.

* Inorden(Izquierda, Ra√≠z, Derecha): Visita el sub√°rbol izquierdo, luego la ra√≠z y finalmente el sub√°rbol derecho. Es √∫til para obtener los elementos de un √°rbol de b√∫squeda binaria en orden ascendente.

* Preorden(Ra√≠z, Izquierda, Derecha): Visita primero la ra√≠z, luego el sub√°rbol izquierdo y finalmente el sub√°rbol derecho. Es √∫til para copiar un √°rbol.

* Postorden(Izquierda, Derecha, Ra√≠z): Visita el sub√°rbol izquierdo, luego el sub√°rbol derecho y finalmente la ra√≠z. Es √∫til para eliminar el √°rbol por completo.

````python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierda = None
        self.derecha = None

# Creamos un √°rbol binario
raiz = Nodo(1)
raiz.izquierda = Nodo(2)
raiz.derecha = Nodo(3)
raiz.izquierda.izquierda = Nodo(4)

# Funci√≥n para recorrer el √°rbol Inorden
def recorrido_inorden(nodo):
    if nodo:
        # 1. Recorre el sub√°rbol izquierdo
        recorrido_inorden(nodo.izquierda)
        # 2. Visita la ra√≠z
        print(nodo.valor, end=" ")
        # 3. Recorre el sub√°rbol derecho
        recorrido_inorden(nodo.derecha)

print("Recorrido Inorden del √°rbol:")
recorrido_inorden(raiz.derecha)
````

# BUSQUEDA SECUENCIAL

EN RESUMEN:  La b√∫squeda secuencial, tambi√©n conocida como b√∫squeda lineal, es el algoritmo de b√∫squeda m√°s simple y directo. Consiste en recorrer una lista o arreglo, elemento por elemento, desde el principio hasta el final, comparando cada uno con el valor que se busca.

#### Analog√≠a üîé
Imagina que est√°s buscando un libro en un estante desordenado de una biblioteca. La √∫nica forma de encontrarlo es revisar cada libro, uno por uno, desde el primer libro del estante hasta el √∫ltimo.

#### Complejidad de Tiempo ‚è±Ô∏è
La b√∫squeda secuencial tiene una complejidad de tiempo de O(n) (lineal). Esto significa que, en el peor de los casos, el n√∫mero de operaciones es directamente proporcional al n√∫mero de elementos (n) en la lista. Si la lista tiene 100 elementos, en el peor de los casos (si el elemento est√° al final o no existe), el algoritmo tendr√° que hacer 100 comparaciones.

### Qu√© podemos hacer?
En Python, no se necesita una funci√≥n especial para una b√∫squeda secuencial. Se puede realizar mediante un c√≥digo de Python simple y b√°sico:

## 1. Uso de un for (bucle)

La forma m√°s com√∫n de implementar una b√∫squeda secuencial es mediante un forbucle. Esto permite iterar por cada √≠ndice y valor de la lista.

````python
def busqueda_secuencial_bucle(mi_lista, objetivo):
    """
    Busca un objetivo en una lista.
    Devuelve el √≠ndice si se encuentra o -1 si no se encuentra.
    """
    # Recorre la lista desde el primer elemento hasta el √∫ltimo
    for i in range(len(mi_lista)):
        # Si el elemento actual es igual al objetivo
        if mi_lista[i] == objetivo:
            return i  # Devuelve el √≠ndice si se encuentra

    # Si el bucle termina sin encontrar el objetivo
    return -1  # Devuelve -1 si no se encontr√≥

mi_lista = [5, 2, 8, 10, 9]

# Llama a la funci√≥n y muestra el resultado
print(f"El √≠ndice es: {busqueda_secuencial_bucle(mi_lista, 20)}")
````

## 2. Uso del in (operador)

Para un enfoque mucho m√°s sencillo, el operador de Python inrealiza una b√∫squeda secuencial en segundo plano. Comprueba la existencia de un elemento en una lista sin necesidad de escribir un bucle.

````python
mi_lista = [5, 2, 8, 1, 9]

if 10 in mi_lista:
    print("est√° en la lista.")
else:
    print(" no est√° en la lista.")
````

Ejemplo 1: B√∫squeda B√°sica en una Lista de N√∫meros üî¢

````python
def busqueda_secuencial_numeros(lista, objetivo):
    """Busca un n√∫mero en una lista. Devuelve el √≠ndice o -1 si no est√°."""
    # Recorremos la lista desde el primer elemento hasta el √∫ltimo
    for i in range(len(lista)):
        # Comparamos el elemento actual con el objetivo
        if lista[i] == objetivo:
            return i  # Si coinciden, devolvemos el √≠ndice

    # Si el bucle termina y no se encontr√≥ el objetivo
    return -1

numeros = [10, 5, 20, 8, 15, 20, 30, 45, 43]

# Buscando un n√∫mero que existe
indice_encontrado = busqueda_secuencial_numeros(numeros, 5)
print(f"El n√∫mero se encuentra en el √≠ndice: {indice_encontrado}")

# Buscando un n√∫mero que no existe
indice_no_encontrado = busqueda_secuencial_numeros(numeros, 80)
print(f"El n√∫mero se encuentra en el √≠ndice: {indice_no_encontrado}")
````

Ejemplo 2: B√∫squeda en una Lista de Cadenas de Texto üìñ

````python
def busqueda_secuencial_palabras(lista_palabras, objetivo):
    """Busca una palabra en una lista. Devuelve el √≠ndice o -1."""
    # Recorremos la lista de palabras
    for i in range(len(lista_palabras)):
        # Comparamos la palabra actual con el objetivo
        if lista_palabras[i] == objetivo:
            return i

    # Si no se encuentra la palabra
    return "no esta"

diccionario_simple = ["manzana", "Pitahaya", "uva", "kiwi", "naranja", "Guaba","Pitahaya"]

# Buscando una palabra que existe
indice_banana = busqueda_secuencial_palabras(diccionario_simple, "Pitahaya")
print(f"La palabra 'banana' se encuentra en el √≠ndice: {indice_banana}")

# Buscando una palabra que no existe
indice_mango = busqueda_secuencial_palabras(diccionario_simple, "mango")
print(f"La palabra 'mango' se encuentra en el √≠ndice: {indice_mango}")
````

# BUSQUEDA BINARIA

#### Analog√≠a üìñ
La mejor analog√≠a para la b√∫squeda binaria es buscar una palabra en un diccionario impreso. No empiezas por la primera p√°gina. En su lugar, abres el libro aproximadamente por la mitad, miras la primera palabra de esa p√°gina y decides si tu palabra est√° antes o despu√©s de esa secci√≥n. Luego, repites el proceso en la mitad restante, descartando la otra mitad.

#### Complejidad de Tiempo ‚è±Ô∏è
La b√∫squeda binaria tiene una complejidad de tiempo de O(logn) (logar√≠tmica), mientras que la b√∫squeda lineal tiene una complejidad de O(n) (lineal).

Esto significa que a medida que el n√∫mero de elementos (n) crece, el tiempo que tarda la b√∫squeda binaria aumenta mucho m√°s lentamente. Por ejemplo, en una lista de un mill√≥n de elementos, la b√∫squeda binaria podr√≠a encontrar un elemento en tan solo 20 pasos, mientras que una b√∫squeda lineal podr√≠a requerir un mill√≥n de pasos.

Ejemplo 1: B√∫squeda de un N√∫mero B√°sico üî¢

````python
def busqueda_binaria(lista_ordenada, objetivo):
    """
    Busca un objetivo en una lista ordenada usando el algoritmo de b√∫squeda binaria.
    Devuelve el √≠ndice del objetivo o -1 si no se encuentra.
    """
    izquierda, derecha = 0, len(lista_ordenada) - 1

    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2
        valor_medio = lista_ordenada[medio]

        if valor_medio == objetivo:
            return medio  # El objetivo se encontr√≥
        elif valor_medio < objetivo:
            izquierda = medio + 1  # Busca en la mitad derecha
        else:
            derecha = medio - 1  # Busca en la mitad izquierda

    return -1  # El objetivo no se encontr√≥

# Lista ordenada para el ejemplo
numeros_ordenados = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]

# Buscando un n√∫mero que existe
indice_encontrado = busqueda_binaria(numeros_ordenados, 23)
print(f"El n√∫mero 23 se encuentra en el √≠ndice: {indice_encontrado}")

# Buscando un n√∫mero que no existe
indice_no_encontrado = busqueda_binaria(numeros_ordenados, 10)
print(f"El n√∫mero 10 se encuentra en el √≠ndice: {indice_no_encontrado}")
````

#### Algoritmo del codigo:
* Buscando 23: El algoritmo comienza con un rango de 0 a 9. En la primera iteraci√≥n, el medio es el √≠ndice 4 (valor 16). Como 16 es menor que 23, el rango se ajusta a 5 a 9. El proceso contin√∫a reduciendo el rango hasta que encuentra el 23.
* Buscando 10: El proceso es similar, pero el n√∫mero 10 nunca se encuentra en la lista. El rango se reduce en cada paso hasta que izquierda se vuelve mayor que derecha, y el bucle termina. El c√≥digo devuelve -1.

````python
def busqueda_binaria_con_pasos(lista_ordenada, objetivo):
    """
    Busca un objetivo en una lista ordenada y cuenta los pasos.
    Devuelve una tupla: (√≠ndice, n√∫mero_de_pasos).
    """
    izquierda, derecha = 0, len(lista_ordenada) - 1
    pasos = 0  # Inicializamos el contador

    while izquierda <= derecha:
        pasos += 1  # Incrementamos el contador en cada iteraci√≥n
        medio = (izquierda + derecha) // 2
        valor_medio = lista_ordenada[medio]

        if valor_medio == objetivo:
            return (medio, pasos)  # Se encontr√≥, devolvemos el √≠ndice y los pasos
        elif valor_medio < objetivo:
            izquierda = medio + 1  # Buscamos en la mitad derecha
        else:
            derecha = medio - 1  # Buscamos en la mitad izquierda

    return (-1, pasos)  # No se encontr√≥, devolvemos -1 y el total de pasos

# Lista de frutas ordenada alfab√©ticamente
frutas_ordenadas = ["ar√°ndano", "banana", "kiwi", "lim√≥n", "mango", "naranja", "uva"]

# Buscando la fruta "kiwi" (que s√≠ existe)
indice_encontrado, pasos_encontrado = busqueda_binaria_con_pasos(frutas_ordenadas, "kiwi")
print(f"La fruta 'kiwi' se encuentra en el √≠ndice: {indice_encontrado}. Pasos: {pasos_encontrado}")

# Buscando la fruta "pera" (que no existe)
indice_no_encontrado, pasos_no_encontrado = busqueda_binaria_con_pasos(frutas_ordenadas, "pera")
print(f"La fruta 'pera' no se encuentra (√≠ndice: {indice_no_encontrado}). Pasos: {pasos_no_encontrado}")
````

#### Ejemplo Visual üîé
Imagina que buscas el n√∫mero 23 en la lista [2, 5, 8, 12, 16, 23, 38, 56, 72, 91].

* **Iteraci√≥n 1:**

izquierda = 0, derecha = 9.

medio = (0 + 9) // 2 = 4.

valor_medio = lista[4] = 16.

Como 16 < 23, el objetivo est√° a la derecha. El c√≥digo ejecuta izquierda = medio + 1, que es 4 + 1 = 5.

* Iteraci√≥n 2: **texto en negrita**

Tu nuevo rango de b√∫squeda es de izquierda = 5 a derecha = 9.

medio = (5 + 9) // 2 = 7.

valor_medio = lista[7] = 56.

Como 56 > 23, el objetivo est√° a la izquierda. El c√≥digo ejecuta derecha = medio - 1, que es 7 - 1 = 6.

* **Iteraci√≥n 3:**

Tu nuevo rango de b√∫squeda es de izquierda = 5 a derecha = 6.

medio = (5 + 6) // 2 = 5.

valor_medio = lista[5] = 23.

Como 23 == 23, se encuentra el objetivo y el algoritmo termina.

Este proceso de ajuste de los l√≠mites izquierda y derecha es lo que le da a la b√∫squeda binaria su eficiencia logar√≠tmica.









Ejemplo 2: B√∫squeda en un Diccionario de Palabras üìñ

````python
def busqueda_binaria_palabra(lista_palabras, palabra_objetivo):
    """
    Busca una palabra en una lista ordenada alfab√©ticamente.
    """
    izquierda, derecha = 0, len(lista_palabras) - 1

    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2
        palabra_media = lista_palabras[medio]

        if palabra_media == palabra_objetivo:
            return medio
        elif palabra_media < palabra_objetivo:
            izquierda = medio + 1
        else:
            derecha = medio - 1

    return -1

# Lista de palabras ordenada alfab√©ticamente
diccionario = ["agua", "casa", "gato", "perro", "sol", "verde"]

# Buscando una palabra que existe
indice_gato = busqueda_binaria_palabra(diccionario, "gato")
print(f"La palabra 'gato' se encuentra en el √≠ndice: {indice_gato}")

# Buscando una palabra que no existe
indice_cielo = busqueda_binaria_palabra(diccionario, "cielo")
print(f"La palabra 'cielo' se encuentra en el √≠ndice: {indice_cielo}")
````

# TABLA HASH

EN RESUMEN: una tabla hash es el tipo de dato integrado conocido como diccionario (dict), que almacena pares de clave-valor de forma eficiente. Utiliza una funci√≥n hash para convertir las claves en √≠ndices num√©ricos, permitiendo operaciones r√°pidas de inserci√≥n, b√∫squeda y eliminaci√≥n de elementos.

### C√≥mo Funcionan ‚öôÔ∏è
Una tabla hash funciona asociando una clave con un valor . Utiliza una funci√≥n hash para convertir la clave en un √≠ndice num√©rico, que luego se usa para almacenar el valor en un arreglo (o array ).

* Funci√≥n Hash : Es el coraz√≥n de la tabla. Toma una clave (como una palabra o un n√∫mero de identificaci√≥n) y la transforma en un n√∫mero entero que servir√° como √≠ndice. Esta funci√≥n debe ser r√°pida y generar √≠ndices distribuidos uniformemente para evitar colisiones.

* √çndice : El n√∫mero generado por la funci√≥n hash.

* Arreglo (Array) : Es la estructura de datos subyacente donde se almacenan los valores. Cada √≠ndice del arreglo corresponde a una "ranura" donde se puede guardar un valor.

### Analog√≠a üóÑÔ∏è
Imagina una biblioteca. La forma tradicional de buscar un libro es recorrer estante por estante (b√∫squeda secuencial). La b√∫squeda binaria ser√≠a como buscar en un cat√°logo ordenado.

Una tabla hash es como un sistema de almacenamiento automatizado. Cuando llega un libro, la biblioteca le asigna un c√≥digo √∫nico (la funci√≥n hash ) que te dice exactamente en qu√© compartimento (el √≠ndice ) guardarlo. Cuando quieras el libro, simplemente le das el c√≥digo a la m√°quina, que te lo entregar√° al instante. No necesitas buscar en ning√∫n otro lugar.

En Python, el tipo de dato dict(diccionario) es una implementaci√≥n de una tabla hash. Su uso es intuitivo y te permite ver el poder de esta estructura.

````python
# Un diccionario es una tabla hash en Python
contactos = {}

# 1. Inserci√≥n (Operaci√≥n O(1))
# Asignamos valores a claves
contactos["Juan Perez"] = "123-456-7890"
contactos["Maria Lopez"] = "987-654-3210"
print("Contactos:", contactos)

# 2. B√∫squeda (Operaci√≥n O(1))
# Buscamos el n√∫mero de tel√©fono de Juan
print("Tel√©fono de Juan:", contactos["Juan Perez"])

# 3. Eliminaci√≥n (Operaci√≥n O(1))
# Eliminamos a Maria de la lista
del contactos["Maria Lopez"]
print("Contactos despu√©s de eliminar:", contactos)
````

Ejemplo 1: Directorio Telef√≥nico üìû

````python
# Creamos un directorio telef√≥nico como una tabla hash
directorio_telefonico = {
    "Ana": "987-654-3210",
    "Luis": "555-123-4567",
    "Sof√≠a": "123-456-7890"
}

# 1. B√∫squeda de un contacto
nombre = "Juan"
if nombre in directorio_telefonico:
    print(f"El tel√©fono de {nombre} es: {directorio_telefonico[nombre]}")
else:
    print(f"El contacto '{nombre}' no se encuentra.")

# 2. Agregar un nuevo contacto
directorio_telefonico["Carlos"] = "999-888-7777"
print("\nDirectorio actualizado:", directorio_telefonico)

# 3. Eliminar un contacto
del directorio_telefonico["Ana"]
print("Directorio despu√©s de eliminar a Ana:", directorio_telefonico)
````
